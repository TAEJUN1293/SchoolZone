총학생수 컬럼 병합.

```python
import pandas as pd
import numpy as np
```

```python
# 어린이집 학생 수 
school = pd.read_csv('C:/Users/wnrrh/desktop/data/전처리/서울특별시 어린이집 학생 수.csv',header=0,encoding='cp949')
```

```python
columns = ['자치구','기관명','운영현황','주소','CCTV 설치대수', '정원', '현원',
          '위도','경도']
school.columns = columns
```

```python
idx = school[school['운영현황'] == '폐지'].index
school = school.drop(idx)
```

```python
idx2 = school[school['현원'] == 0].index 
school = school.drop(idx2)
```

```python
school = school.reset_index()
school
```

```python
school = school.drop(['index','CCTV 설치대수','정원','위도','경도'],axis=1, inplace = False)
```

```python
data_list = school['주소'].str.split()
school['시'] = data_list.str.get(0)
school['구'] = data_list.str.get(1)
school['도로명'] = data_list.str.get(2)
school['상세'] = data_list.str.get(3)
school
```

```python
school.drop(['주소'],axis=1,inplace=False)
```

```python
school['주소'] = school['시']+" "+school['구']+" "+school['도로명']+" "+school["상세"]
school
```

```python
school = school.drop(['시','구','도로명','상세','운영현황'],axis=1, inplace=False)
school
```

```python
school = school.drop(['level_0'],axis=1,inplace=False)
```
