# 사고 데이터 차종 전처리

```python
columns = ['사고일시','시군구','사망자수','중상자수','경상자수','부상신고자수',
          '가해운전자 차종','피해운전자 차종','피해운전자 연령','시간대','행정동',
          '시설명']
accident.columns = columns
```

```python
accident['가해운전자 차종'].unique()
```

사망사고 비율을 기준으로 한 가해운전자 차종 수치화
https://www.sisain.co.kr/news/articleView.html?idxno=45592 기준

해당 차종으로 가해시 사망 수준이 높을수록 값을 크게 처리

```python
accident['가해운전자 차종'].replace('자전거','3',inplace = True)
accident['가해운전자 차종'].replace('기타불명','999',inplace = True)
accident['가해운전자 차종'].replace('이륜','2',inplace = True)
accident['가해운전자 차종'].replace('원동기','4',inplace = True)
accident['가해운전자 차종'].replace('승용','5',inplace = True)
accident['가해운전자 차종'].replace('승합','7',inplace = True)
accident['가해운전자 차종'].replace('화물','6',inplace = True)
accident['가해운전자 차종'].replace('건설기계','8',inplace = True)
accident['가해운전자 차종'].replace('특수','1',inplace = True)
accident['가해운전자 차종'].replace('개인형이동수단(PM)','3',inplace = True)
```
