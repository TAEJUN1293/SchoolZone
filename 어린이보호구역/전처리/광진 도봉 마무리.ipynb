```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('C:/Users/wnrrh/desktop/data/현황/광진도봉.xlsx',header = 0)
```

```python
columns = ['대상시설명','소재지도로명주소','자치구','시설종류','구역지정수',
          'CCTV설치여부','CCTV 수','시설물 도로표지판 수','시설물 과속방지턱 수',
           '보호구역도로폭','위도','경도']
df.columns = columns
```

```python
df.drop(['자치구'],axis=1, inplace = True)
```

```python
df = df.rename(columns={'소재지도로명주소':'주소'})
df
```

```python
time = pd.DataFrame([], columns = ['시간대'] + list(df.keys())[:11] + list(df.keys())[13:])
for i in range(len(df)):
    for t in range(1,25):
        time.loc[len(time)] = ([t] + list(df.iloc[i,:11]) + list(df.iloc[i,13:]))
time = time.iloc[:,[1,0]+[i for i in range(2, len(time.keys()))]]
time
```

```python
df = time
df['시간대'] = df['시간대'].apply(lambda x: "{:0>2d}".format(x))

df
```

```python
df['시간대'] = df['시간대'].map('{}시'.format)
```
