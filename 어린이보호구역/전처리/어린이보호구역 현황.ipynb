```python
df1 = pd.read_excel('C:/Users/wnrrh/desktop/data/현황/성동영등포구.xlsx',header=0)
df2 = pd.read_excel('C:/Users/wnrrh/desktop/data/현황/강남관악 정리.xlsx',header=0)
```

```python
columns = ['대상시설명','주소','자치구','시설종류','구역지정수','CCTV설치여부',
          'CCTV 수','시설물 도로표지판 수','시설물 과속방지턱 수',
           '보호구역도로폭','위도','경도']
df1.columns = columns
df2.columns = columns
```

```python
df1.drop(['자치구'],axis=1, inplace = True)
```

```python
df1
```

```python
time = pd.DataFrame([], columns = ['시간대'] + list(df1.keys())[:11] + list(df1.keys())[13:])
for i in range(len(df1)):
    for t in range(1,25):
        time.loc[len(time)] = ([t] + list(df1.iloc[i,:11]) + list(df1.iloc[i,13:]))
time = time.iloc[:,[1,0]+[i for i in range(2, len(time.keys()))]]
```

```python
df1 = time
df1['시간대'] = df1['시간대'].apply(lambda x: "{:0>2d}".format(x))
```

```python
df1['시간대'] = df1['시간대'].map('{}시'.format)
```
